```python
# Load the environment variables from .env file
load_dotenv()

# Now you can access the environment variable just like before
client = OpenAI(
    api_key=os.environ.get("OPENAI_APIKEY"), 
)
```

```python
# Define a function to interact with the model
def generate_text(prompts):
    try:

        response = client.chat.completions.create(
            model="gpt-4-turbo",
            messages=prompts,
            response_format={
                "type": "text"
            },
            temperature=0.1,
            #max_tokens=16384,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0
        )
        return response
    except Exception as e:
        return f"Error: {e}"
```

```python
prompts = [
    { "role":"assistant" , "content":"#### ISTRUZIONI\n analizza la determina dirigenziali e poi rispondi a delle domande"+determina},
    { "role":"assistant" , "content":"Rispondi punto per punto se le seguenti normative sono rispettate"+question},

]
```

```python
result = generate_text(prompts)
print("Generated Response:", result)
```
